```python
#importing the libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
#load the dataset
ir=load_iris()
iris=pd.DataFrame(ir.data,columns=ir.feature_names)
iris.head()
```

```python
x=iris.iloc[:,:].values
```

```python
#The Elbow method for finding K value
kv=range(1,10)
ssv=[]
for k in kv:
    km=KMeans(n_clusters=k)
    km.fit(x)
    ssv.append(km.inertia_)
```

```python
ssv
```

```python
#plotting the results to observe the elbow
plt.title('The Elbow method ')
plt.xlabel('no of clusters')
plt.ylabel('ssv values')
plt.plot(kv,ssv)
```

```python
#traning the model 
km=KMeans(n_clusters=3)
y_pred=km.fit_predict(x)
y_pred
```

```python
#visualizing the clusters
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='purple',label='Iris-setosa')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='yellow',label='Iris-versicolour')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='green',label='Iris-virginica')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='*',c='black',label='centroids')
plt.legend()
```
